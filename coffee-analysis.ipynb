## EDA

```python
df.columns = ['Category', 'Value']

print(df.head())
```

```python
df['Value'] = pd.to_numeric(df['Value'], errors='coerce')
```

```python
df.info()
df.describe()
```

```python
df.columns = df.iloc[0]

df = df[1:]
```

```python
df.columns = ['Customer_Name', 'Sales']
```

```python
df['Sales'] = pd.to_numeric(df['Sales'])
```

```python
df.head()
df.info()
```

```python
plt.figure(figsize=(12,6))

plt.bar(df['Customer_Name'], df['Sales'], color='brown')

plt.title("Coffee Sales by Customer")
plt.xlabel("Customer Name")
plt.ylabel("Sales")

plt.xticks(rotation=90)

plt.show()
```

```python
top10 = df.sort_values(by='Sales', ascending=False).head(10)

plt.figure(figsize=(10,6))

sns.barplot(x='Sales', y='Customer_Name', data=top10, palette='magma')

plt.title("Top 10 Coffee Customers")

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.histplot(df['Sales'], bins=20, color='darkgreen', kde=True)

plt.title("Coffee Sales Distribution")

plt.show()
```

```python
plt.figure(figsize=(6,6))

sns.boxplot(y=df['Sales'], color='orange')

plt.title("Sales Distribution Boxplot")

plt.show()
```

```python
top5 = df.sort_values(by='Sales', ascending=False).head(5)

plt.figure(figsize=(8,8))

plt.pie(top5['Sales'],
        labels=top5['Customer_Name'],
        autopct='%1.1f%%',
        colors=['brown','orange','gold','darkred','peru'])

plt.title("Top 5 Customer Sales Share")

plt.show()
```
